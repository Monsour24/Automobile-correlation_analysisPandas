```python
sorted_df_price = df.sort_values('price', ascending=False)

sorted_df_price.head(1)
```

```python
# Use boolean indexing to select rows that meet the condition
Toyota_automobile = (df['company'] == 'toyota') 

Toyota_automobile = df[Toyota_automobile]
Toyota_automobile
```

```python
# select the column for company
count_company = df['company'].value_counts()

# display the resulting counts
count_company
```

```python
# group the data by 'company', and find the index of the maximum value in 'price' for each group
max_price_idx = df.groupby('company')['price'].idxmax()

# use the index to retrieve the corresponding values of 'company', 'body_style', and 'price'
max_price_cars = df.loc[max_price_idx, ['company', 'body-style', 'price']]

# display the resulting dataframe
max_price_cars
```
